```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scienceplots
from scipy import stats
```

```python
m = 0.015   # [kg]
r = 6.9 * 1e-3 # [mm]
```

```python
def position(t, omega, X0, A, Beta, Delta):
    return X0 + A * np.exp(0.5*-Beta * t /m) * np.cos(omega * t + Delta)

def fit_curve(data, function, initial_guess, plot=False):
    x = data[0]; y = data[1]

    # Fit the data to the model
    popt, pcov = curve_fit(function, x, y, p0=initial_guess)

    # Print the optimized parameters
    print("Optimized Parameters:")
    for param in popt:
        print(f'{param=:.4f}')

    xx = np.linspace(x[0],x[-1],1000)
    fitted_curve = function(xx, *popt)

    if plot:
    # Plot the original data and the fitted curve
        plt.figure(figsize=(10, 6))
        plt.scatter(x, y, label='Original Data')
        plt.plot(xx, fitted_curve, color='red', label='Fitted Curve')
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.title('Fitted Curve with Optimized Parameters')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
df = pd.read_excel('data.xlsx',sheet_name='kappa')
x_col = 1
t_col = 0

ti = 19
tf = 106

x = df.iloc[ti:tf,x_col].to_numpy()
t = df.iloc[ti:tf,t_col].to_numpy()

data = [t,x]

fit_curve(data, position, [.01, 50, -50, 0, 0], plot=True)
```
